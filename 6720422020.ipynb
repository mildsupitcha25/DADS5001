```python
import pandas as pd
import plotly.express as px
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
df.head()
```

```python
df.info()
```

```python
df['item_price'] = df['item_price'].str.replace('$', '', regex=False).astype(float)
display(df.head())
```

```python
top_10_items = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(10)
print("Top 10 items by quantity sold:")
print(top_10_items)
```

```python
nan_count = df.isnull().sum()
print(f"Number of NaN values: {nan_count}")
```

```python
df['item_name'].unique()
```

```python
df['choice_description'].unique()
```

```python
df['item_price'].unique()
```

```python
df['revenue'] = df['quantity'] * df['item_price']
display(df.head())
```

```python
item_revenue = df.groupby('item_name')['revenue'].sum()
```

```python
top_10_revenue_items = item_revenue.sort_values(ascending=False).head(10)
print("Top 10 items by revenue:")
print(top_10_revenue_items)
```

```python
top_10_revenue_items_df = top_10_revenue_items.reset_index()
top_10_revenue_items_df.columns = ['item_name', 'revenue']

fig = px.bar(top_10_revenue_items_df, x='item_name', y='revenue', title='Top 10 Items by Revenue')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
item_qty = df.groupby('item_name')['quantity'].sum()
```

```python
top_10_qty_items = item_qty.sort_values(ascending=False).head(10)
print("Top 10 items by quantity:")
print(top_10_qty_items)
```

```python
top_10_qty_items_df = top_10_qty_items.reset_index()
top_10_qty_items_df.columns = ['item_name', 'quantity']

fig = px.bar(top_10_qty_items_df, x='item_name', y='quantity', title='Top 10 Items by quantity')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
fig = px.scatter(df,
                x='item_name',
                y='revenue',
                color = 'item_price',
                size = 'quantity',
                title='item VS revenue')
fig.show()
```

```python
# Calculate total revenue for each item
item_revenue = df.groupby('item_name')['revenue'].sum()

# Get the top 10 items by revenue
top_10_revenue_items = item_revenue.sort_values(ascending=False).head(10)

# Filter the original DataFrame to include only the top 20 items
df_top_10_revenue = df[df['item_name'].isin(top_10_revenue_items.index)]

# Create the scatter plot
fig = px.scatter(df_top_10_revenue,
                x='item_name',
                y='revenue',
                color = 'item_price',
                size = 'quantity',
                title='Top 10 Items by Revenue')
fig.show()
```
